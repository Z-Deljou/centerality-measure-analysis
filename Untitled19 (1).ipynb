visualizing the graph with matplotlib

```python
def visualize_the_graph(G,plt):
  plt.figure(figsize=(8, 6))
  nx.draw(G, with_labels=True, node_size=300, node_color='skyblue', font_size=8)
  plt.title("Graph Visualization")
  return plt
```

finding avg shortest path length,diameter,avg degree,number of edges,number of nodes,avg shortest path length with using networkx library

```python
def metrics_for_graph(data):
  G = nx.from_pandas_edgelist(data,source= "id1", target="id2")
  num_nodes = G.number_of_nodes()
  num_edges = G.number_of_edges()
  avg_degree = sum(dict(G.degree()).values()) / num_nodes
  density = nx.density(G)
  diameter = nx.diameter(G)
  avg_shortest_path_length = nx.average_shortest_path_length(G)
  return avg_shortest_path_length,diameter,avg_degree,num_edges,num_nodes,avg_shortest_path_length, density, diameter
```

finding clustering coeficient1 with using networkx library for node 1

```python
def clustering_coef_1(G):
  local_clustering = nx.transitivity(G)
  return local_clustering
```

measuring centrality metrics with using networkx lib

```python
def centrality_metrics(G):
  local_clustering_coefficients = [nx.clustering(G, node) for node in G.nodes()]
    # Calculate the average of the local clustering coefficients
  global_clustering = sum(local_clustering_coefficients) / len(local_clustering_coefficients)
  average_clustering_coefficient = nx.average_clustering(G)
  # Betweenness Centrality
  betweenness_centrality = nx.betweenness_centrality(G)
  # Closeness Centrality
  closeness_centrality = nx.closeness_centrality(G)
  # Degree Centrality
  degree_centrality = nx.degree_centrality(G)
  # Network Centralization
  # Calculate the network centralization based on degree centrality
  degree_values = list(degree_centrality.values())
  network_size = len(G)
  max_possible_sum = (network_size - 1) * (network_size - 2)
  network_centralization = (max_possible_sum - sum(degree_values)) / max_possible_sum
  return average_clustering_coefficient,betweenness_centrality,closeness_centrality,degree_centrality,network_centralization
```

finding the centrality nodes such as:top three betweenness,top three closeness,top three degree for detecting important nodes

```python
def centrality_nodes(G):
  betweenness_centrality = nx.betweenness_centrality(G)
  # Closeness Centrality
  closeness_centrality = nx.closeness_centrality(G)
  # Degree Centrality
  degree_centrality = nx.degree_centrality(G)
  top_three_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:3]
  top_three_closeness = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:3]
  top_three_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:3]
  return top_three_betweenness,top_three_closeness,top_three_degree
```

```python
def main():
    plot = visualize_the_graph(G,plt)
    plot.show()
    avg_shortest_path_length,diameter,avg_degree,num_edges,num_nodes,avg_shortest_path_length, density, diameter =metrics_for_graph(data)
    print("# Nodes:", num_nodes)
    print("# Edges:", num_edges)
    print("Average Degree:", avg_degree)
    print("Density:", density)
    print("Diameter:", diameter)
    print("Average Shortest Path Length:", avg_shortest_path_length)
    clustering_node_1 = clustering_coef_1(G)
    print("Clustering Coefficient for Node 1:", clustering_node_1)
    average_clustering_coefficient,betweenness_centrality,closeness_centrality,degree_centrality,network_centralization=centrality_metrics(G)
    print("Average Clustering Coefficient:", average_clustering_coefficient)
    print("Betweenness Centrality:", betweenness_centrality)
    print("Closeness Centrality:", closeness_centrality)
    print("Degree Centrality:", degree_centrality)
    print("Network Centralization:", network_centralization)
    top_three_betweenness,top_three_closeness,top_three_degree=centrality_nodes(G)
    print("Top three nodes with the highest Betweenness Centrality:")
    for node, centrality in top_three_betweenness:
        print(f"Node {node}: {centrality}")
    print("\nTop three nodes with the highest Closeness Centrality:")
    for node, centrality in top_three_closeness:
        print(f"Node {node}: {centrality}")

    print("\nTop three nodes with the highest Degree Centrality:")
    for node, centrality in top_three_degree:
        print(f"Node {node}: {centrality}")
main()
```
